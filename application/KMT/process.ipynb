```python
import numpy as np
import pandas as pd
import os
eventname = 'KB180800'
obs = ['KMTA01', 'KMTA42', 'KMTC01', 'KMTC42', 'KMTS01', 'KMTS42']
path = './%s'%(eventname)
```

```python
files = os.listdir(path)
data = [np.loadtxt(os.path.join(path, x)) for x in files]
print(data)

# kmta01 = np.loadtxt('KB-18-0800-kmta01.dat')
# kmta42 = np.loadtxt('KB-18-0800-kmta42.dat')
# kmtc01 = np.loadtxt('KB-18-0800-kmtc01.dat')
# kmtc42 = np.loadtxt('KB-18-0800-kmtc42.dat')
# kmts01 = np.loadtxt('KB-18-0800-kmts01.dat')
# kmts42 = np.loadtxt('KB-18-0800-kmts42.dat')
# data = [kmta01, kmta42, kmtc01, kmtc42, kmts01, kmts42]

# Tel,Filter,HJD,mag,e_mag,mag_aligned,e_mag_aligned
# date,modmag,imag,imerr
```

```python
Tel = np.hstack([np.full((1, len(ob)), obs[i]) for i,ob in enumerate(data)])[0]
Filter = np.full((sum(len(x) for x in data)), 'I')
HJD = np.hstack([ob[:, 0] for ob in data])
mag_aligned = np.hstack([ob[:, 2] for ob in data])
e_mag_aligned = np.hstack([ob[:, 3] for ob in data])
mag = np.zeros((sum(len(x) for x in data)))
e_mag = np.zeros((sum(len(x) for x in data)))
```

```python
data_aligned = pd.DataFrame({'Tel':Tel, 'Filter':Filter, 'HJD':HJD, 'mag':mag, 'e_mag':e_mag, 'mag_aligned':mag_aligned, 'e_mag_aligned':e_mag_aligned})
data_aligned.to_csv('./%s/%s_aligned.csv'%(eventname, eventname), index=False, sep=',')
```
