```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = []
for root, dirs, files in os.walk('./KMT-2020-BLG-0414/'):
    for f in files:
        if 'V' not in f:
            data.append(np.loadtxt(f'./KMT-2020-BLG-0414/{f}'))
data = np.vstack(data)
```

```python
data = data[np.argsort(data[:, 0])]
```

```python
t0 = 9041.55526; te = 90.53
plt.xlim(t0-2*te, t0+2*te)
plt.scatter(data[:, 0], data[:, 3])
```

```python
a = data[(data[:, 0] > t0-2*te) * (data[:, 0] < t0+2*te)]
```

```python
a.shape
```

```python
a[:, 0] = (a[:, 0] - t0)/(4*te)
```

```python
plt.scatter(a[:, 0], a[:, 3], s=0.2)
```

```python
np.save('./0414.npy', a[:, [0, 3]])
```
